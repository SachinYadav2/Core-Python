# Break statement
- eska use karke aap loop se bahar(out) aa skte h.
https://cdn.programiz.com/sites/tutorial2program/files/flowchart-break-statement.jpg

```python
while 5>2:
    if 2>1:
        print("Yes All true")
        break  # ye ab niche nhi jaega
    if 4>2:
        print("Print Karo muje")
print("Rest of the code")
```

```python
for i in range(10):
    print(i)
```

```python
for i in range(10):
    
    print(i)
    if i==5:
        break  # Jha break lag gya uske bad ka pura code exexute nhi hota h jo bhi  loop me likha hoga
    print(" if  bahar h ham  se ")
    
print("Rest of the code")
```

# Continue 
- eksa use krke ham kise particular statement ko skip karte h
https://cdn.programiz.com/sites/tutorial2program/files/continue-statement-flowchart.jpg

```python
while 5>2:
    if 2>1:
        print("Yes All true")
        continue  # infnite time tak chalega ye loop

print("Rest of the code")
```

```python
# Always continue ke bad hi ham apne stateemnt ko print krwa skte h

for i in range(10):
    
    if i==5:
        continue  # ye loop only hmara es condtions ko skip krega
    print(i)

    print(" if  bahar h ham  se ")
    
print("@ ---------Rest of the code@")
```
